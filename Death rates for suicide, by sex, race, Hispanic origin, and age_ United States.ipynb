# Data wrangling

The following are the necessary corrections for the dataset:

1) Drop the column FLAG, AGE NUM, YEAR NUM, STUB_LABEL_NUM, STUB_NAME_NUM, UNIT_NUM)
2) Standardize the AGE column
3) Split STUB_LABEL, UNIT
4) Fill in the null vlue for estimates.
5) Convert YEAR datatype to date datatype
6) Convert age to int

1) Dropping the FLAG, AGE_NUM, YEAR_NUM, STUB_LABEL_NUM, STUB_NAME_NUM, UNIT_NUM columns

```python
# To copy the dataset

db = df.copy()
```

```python
# Droping columns

db.drop(columns = ["FLAG", "AGE_NUM", "YEAR_NUM", "STUB_LABEL_NUM", "UNIT_NUM"], axis = 1, inplace = True)
```

```python
db.drop(columns = "STUB_NAME_NUM", axis = 1, inplace = True) 
```

```python
db.head(10)
```

```python
db.tail(10)
```

```python
db["STUB_LABEL"].value_counts()
```

2) Standardize the AGE column

```python
# getting the unique values in AGE column

db["AGE"].value_counts()
```

```python
#To filter and drop the "All ages" in the AGE column

for x in db.index:
    
    if db.loc[x, "AGE"] == "All ages":
        
        db.drop(x, inplace = True)
        
db.head(5)        
```

```python
db.info()
```

```python
db.head(20)
```

```python
db["AGE"].value_counts()
```

```python
# Reseting the index

db.reset_index(drop=True)
```

```python
# Replacing "65 and over" with "65-84" and "85 and over" with "85-100"

db["AGE"] = db["AGE"].str.replace("65 years and over", "65-84") 
```

```python
db["AGE"] = db["AGE"].str.replace("85 years and over", "85-100")
```

```python
# Confirming changes

db["AGE"].value_counts()
```

```python
# Removing the "years" from the age dataset

db[["Age Range", "year"]] = db["AGE"].str.split(" ", 2, expand = True)
```

```python
#Checking for changes

db.head(5)
```

```python
# Dropping the "AGE" and "year" column

db.drop(columns= ["year", "AGE"], axis = 1, inplace = True)
```

```python
# Confirming changes

db.head(10)
```

```python
#Checking unique value for age

db["Age Range"].value_counts()
```

Checking for Null values in the "Estimate column"


```python
#Checking null values

db["ESTIMATE"].isnull().sum()
```

```python
# Filling in the missing values using the mean of the column

db.fillna(db["ESTIMATE"].mean(), inplace = True)
```

```python
#Confirming changes

db.info()
```

```python
#Checking the first 10 rows

db.head(10)
```

Standardizing the YEAR column 

```python
#Changing the data type 

db["YEAR"] = pd.to_datetime(db["YEAR"])
```

```python
#Checking for changes

db.info()
```
